```python
import pandas as pd
```

```python
# Lê as planilhas do excel
df1 = pd.read_excel("Aracaju.xlsx")
df2 = pd.read_excel("Fortaleza.xlsx")
df3 = pd.read_excel("Natal.xlsx")
df4 = pd.read_excel("Recife.xlsx")
df5 = pd.read_excel("Salvador.xlsx")
```

```python
# concatena todas as planilhas
df = pd.concat([df1, df2, df3, df4, df5])
```

```python
df.head()
```

```python
# transformando a coluna de data em tipo inteiro
df["Data"] = df["Data"].astype("int64")
```

```python
df.dtypes
```

```python
# transformando a coluna Data em data
df["Data"] = pd.to_datetime(df["Data"])
```

```python
df.dtypes
```

```python
# Agrupa por ano e exibe o somatorio de vendas por ano
df.groupby(df["Data"].dt.year)["Vendas"].sum()
```

```python
# Criando uma nova coluna com o ano
df["Ano Venda"] = df["Data"].dt.year
```

```python
df.sample(5)
```

```python
# extraindo o mês e o dia
df["Mês Venda"], df["Dia Venda"] = (df["Data"].dt.month, df["Data"].dt.day)
```

```python
df.sample(5)
```

```python
# retorna a data mais antiga
df["Data"].min()
```

```python
# calcula a diferença de dias
df["Diferença Dias"] = df["Data"] - df["Data"].min()
```

```python
df.sample(5)
```

```python
# criando a coluna de trimestre
df["Trimestre Venda"] = df["Data"].dt.quarter
```

```python
df.sample(5)
```

```python
# Filtrando as vendas de 2019 do mês de março
vendas_marco_19 = df.loc[(df["Data"].dt.year == 2019) & (df["Data"].dt.month == 3)]
```
